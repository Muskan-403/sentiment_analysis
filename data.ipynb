```python
eg = df['Text'][40]
print(eg)
```

```python
tokens=nltk.word_tokenize(eg)
tokens[:10]
```

```python
tagged=nltk.pos_tag(tokens)
tagged[:10]
```

```python
entities= nltk.chunk.ne_chunk(tagged)
entities.pprint()
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
sia = SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores('I am Happy!')
```

```python
sia.polarity_scores('This is worst thing ever.')
```

```python
sia.polarity_scores(eg)
```

```python
res ={}
for i, row in tqdm(df.iterrows(), total=len(df)):
    text = row['Text']
    myid = row['Id']
    res[myid] = sia.polarity_scores(text)
```

```python
vader=pd.DataFrame(res).T
vader=vader.reset_index().rename(columns={'index':'Id'})
vader= vader.merge(df, how='left')
```

```python
ax = sns.barplot(data=vader, x='Score', y='compound', palette=['#FF5733', '#33C1FF', '#75FF33', '#E833FF', '#BB34FF'], hue='Score', legend=False)
ax.set_title('Compound Score by Amazon Star Review')
plt.show()
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,5))
sns.barplot(data=vader,x='Score',y='pos', ax=axs[0],color='blue')
sns.barplot(data=vader,x='Score',y='neu', ax=axs[1], color='green')
sns.barplot(data=vader,x='Score',y='neg', ax=axs[2])
axs[0].set_title('Positive')
axs[1].set_title('Neutral')
axs[2].set_title('Negative')
plt.show()
```
